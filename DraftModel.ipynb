```python
# Simulating 100 Snake Drafts and each time the model is 
# assigned a random  draft pick to see how overtime the model 
# performs againstbthe baseline

num_simulations = 100  # number of simulated drafts
num_teams = 12 # typical number of teams in each league
rounds = 15 #each team drafts one player per round for a total of 15 players total on the roster

model_total_points = [] # Initializing to track points the model gets
baseline_total_points = [] #initializing to track the points the baseline teams get
model_draft_positions = []  # track where the model drafts (0 = first pick)

for sim in range(num_simulations):
    # Randomly assign model’s pick position (0–11)
    model_team = np.random.randint(0, num_teams)
    # Track which draft pick the model gets assigned for each iteration
    model_draft_positions.append(model_team + 1)

    # Get player pool and reset drafted status to everyone undrafted
    player_pool = df_sorted_24.copy().sort_values(by='ADP', ascending=True).reset_index(drop=True)
    player_pool['Drafted'] = False

    # Initilizing each team to be empty
    teams = {i: [] for i in range(num_teams)}

    for round_num in range(rounds):
        # Determining which direction to iterate through the teams (1->12 or 12->1)
        if round_num % 2 == 0:
            draft_order = list(range(num_teams))
        else:
            draft_order = list(reversed(range(num_teams)))

        for pick in draft_order:
            if pick == model_team:
                # Draft based on predicted ADP
                avaliable = player_pool[player_pool['Drafted']==False].sort_values(by = "Predicted_ADP", ascending=True)
                draft_pick = avaliable.iloc[0]
            else:
                #draft based on actual ADP
                avaliable = player_pool[player_pool['Drafted']==False].sort_values(by = "ADP", ascending=True)
                draft_pick = avaliable.iloc[0]
            # Mark drafted player as drafted in player pool
            player_pool.loc[player_pool['Player'] == draft_pick['Player'], 'Drafted'] = True
            # Add draft pick to team's roster
            teams[pick].append(draft_pick)

    # Compute total fantasy points for the model's team
    model_points = pd.DataFrame(teams[model_team])['ProjectedPoints'].astype(float).sum()
    model_total_points.append(model_points)

    # Computing average points across baseline teams for this simulation
    baseline_points = [
        pd.DataFrame(teams[i])['ProjectedPoints'].astype(float).sum()
        for i in range(num_teams) if i != model_team
    ]
    baseline_total_points.append(np.mean(baseline_points))

# Compare average points for model vs baseline approach
avg_model_points = np.mean(model_total_points)
avg_baseline_points = np.mean(baseline_total_points)

# Printing out results:
print("After " + str(num_simulations) + " simulated drafts:")
print(f"Average Model Team Total Points: " + str(round(avg_model_points,2)))
print(f"Average ADP-Based Team Total Points: " + str(round(avg_baseline_points,2)))
print(f"Model advantage per draft: " + str(round(avg_model_points - avg_baseline_points,2)) + " points")

# Showing distribution of which draft pick the model was assigned
plt.figure(figsize=(8,4))
plt.hist(model_draft_positions, bins=np.arange(1, num_teams+2)-0.5, rwidth=0.8, color='skyblue', edgecolor='black')
plt.xticks(range(1, num_teams+1))
plt.xlabel("Draft Position")
plt.ylabel("Frequency")
plt.title("Distribution of Model’s Draft Slot Across " + str(num_simulations) + " Simulations")
plt.show()

```
